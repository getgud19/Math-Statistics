```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

b = np.arange(0.2, 3.2, 0.2)
d = np.arange(0.1, 1.0, 0.1)

B, D = np.meshgrid(b, d)
nu = np.sqrt( 1 + (2*D*B)**2 ) / np.sqrt( (1-B**2)**2 + (2*D*B)**2)

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(B, D, nu)
plt.xlabel('b')
plt.ylabel('d')
plt.show()
```

```python
from sympy import symbols, Eq, solve

x, y, c = symbols('x y c')

# Define the differential equation
eq = Eq(x*(x-1)*y.diff(x) + y*x.diff(x), x**2*(2*x-1))

# Define the solution
sol = y - (x/(x-1))*(x**2 - x + c)

# Substitute the solution into the differential equation
eq_sol = eq.subs(y, sol)

# Solve for c
c_sol = solve(eq_sol, c)

# Substitute the solution for c back into the solution for y
y_sol = sol.subs(c, c_sol[0])

y_sol
```
